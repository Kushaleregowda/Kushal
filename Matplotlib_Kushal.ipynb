```python
import numpy as np
t=np.array([1,2,3,4,5,6,7])
print(t)
```

```python
plt.plot(t,t**2, 'b--', label='^2')
plt.plot(t,t**2.4, 'r--', label='^2.2')
plt.plot(t,t**2.6,'g--', label='^2.5')
plt.xlabel('Numbers', color='g')
plt.ylabel('Square',color='b')
plt.title('Square of a numbers',color='y')
plt.grid()
plt.legend()
plt.show()
```

```python
x1=[1,2,3,4]
y1=[5,6,7,8]
x2=[1,3,7,9]
y2=[2,4,6,8]
lines=plt.plot(x1,x2,y1,y2)

#using keyword arguments
    
plt.setp(lines[0],color='r', linewidth=2.0)
#plt.setp(line[2]), color='b', linewidth=1.0)
plt.grid()
plt.show()
```

# Multiple plots

```python
t1=np.arange(0.1,1.0,0.1)
t2=np.arange(0.5,5,0.6)
plt.figure(1)
```

```python
plt.subplot(211)
plt.grid()
plt.plot(t1, np.f(t1), 'b-')
```

```python
plt.plot(212)
plt.plot(t2, np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3,4])
plt.subplot(212)
plt.plot([5,6,7,1])
plt.show()
```
